```
co_total_confirmed_cases = co.groupby(["Name of State / UT"]).Active.agg(sum)
co_total_confirmed_cases = pd.DataFrame(co_total_confirmed_cases)
co_total_confirmed_cases.Active =co_total_confirmed_cases .Active.map(lambda p: round(p)) #for graph plotting.
co_total_confirmed_cases.plot(ylabel='total confirmed cases', figsize=(10,5))#setting for graph size
co_total_confirmed_cases.T
```

```
co_Deaths = co.groupby(["Name of State / UT"]).Deaths.agg(sum)
co_Deaths = pd.DataFrame(co_Deaths)
co_Deaths.Actived =co_Deaths.Deaths.map(lambda p: round(p)) #for graph plotting.
co_Deaths.plot(ylabel='Deaths', figsize=(10,5))#setting for graph size
co_Deaths.T
```
